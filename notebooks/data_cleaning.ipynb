# 1. Loading datasets to create dataframe with yearly data


```python
def read_file(filename):
	return pd.read_excel(filename)

# Get all filesnames under parent folder data
import os
files = os.listdir('../data')
files = [f for f in files if f.endswith('.xlsx')]

# Create a dataframe for each file, with the same name as the file
for f in files:
    globals()[f.split('.')[0]] = read_file('../data/' + f)
    
# Create a dictionary with the dataframes
data = {f.split('.')[0]: globals()[f.split('.')[0]] for f in files}

# Print all the dataframe names
print(data.keys())
```
